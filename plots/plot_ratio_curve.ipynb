```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
results = pd.read_csv('plot_ratios_us_crime')
```

```python
results.columns
```

```python
plt.xlabel('Ratio')
plt.ylabel('Result')
plt.title('Curvas Ratios')

for c in results.columns:
    if c != 'Ratio':
        plt.plot(results['Ratio'], results[c])

plt.ylim(0, 1.05)
plt.legend(results.columns[1:], loc=4)
plt.savefig('plot_ratios.svg')
plt.show()
plt.close()
```

```python
plt.xlabel('Ratio')
plt.ylabel('Result')
plt.title('Curvas Ratios')

plt.plot(results['Ratio'], results['TPR'], 'b^--')
plt.plot(results['Ratio'], results['TNR'], 'y^--')
plt.plot(results['Ratio'], results['TPRxTNR'], 'r-o')

plt.ylim(0.5, 1.05)
plt.legend(results.columns[1:], loc=4)
plt.savefig('plot_ratios.svg')
plt.show()
plt.close()
```
