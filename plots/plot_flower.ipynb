```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
results = pd.read_csv('res_e4_gmean.csv', sep='\t')
```

```python
results.columns
```

```python
plt.xlabel('Round')
plt.ylabel('G-mean')
plt.title('Resultados FLOWER modelos locales')

for c in results.columns:
    if c != 'Ro.' and c != 'VAE Fed' and c != 'GAN Fed':
        plt.plot(results['Ro.'], results[c])

plt.ylim(0.88, 0.965)
plt.legend(results.columns[1:6])
plt.savefig('plot_results_flower_1.svg')
plt.show()
plt.close()
```

```python
plt.xlabel('Round')
plt.ylabel('G-mean')
plt.title('Resultados FLOWER modelos federados')

for c in results.columns:
    if c != 'Ro.' and c != 'Base Line' and c != 'SMOTE' and c != 'ROS':
        plt.plot(results['Ro.'], results[c])

plt.ylim(0.93, 0.965)
plt.legend(results.columns[4:], loc='lower center')
plt.savefig('plot_results_flower_2.svg')
plt.show()
plt.close()
```
