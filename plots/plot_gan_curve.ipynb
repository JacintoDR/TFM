```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
results = pd.read_csv('gan_loss_abalone.csv')
results.columns
```

```python
results
```

```python
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Curvas entrenamiento GAN')

for c in results.columns:
    if c != 'Epoch':
        plt.plot(results['Epoch'], results[c])

plt.ylim(0, 2.3)
plt.legend(results.columns[1:], loc=1)
plt.savefig('plot_gan_losses1.svg')
plt.show()
plt.close()
```

```python
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Curvas entrenamiento GAN')

plt.plot(results['Epoch'], results['Discriminator'], 'b--')
plt.plot(results['Epoch'], results['Generator'], 'g-')

plt.ylim(0, 2.3)
plt.legend(results.columns[1:], loc=1)
plt.savefig('plot_gan_losses2.svg')
plt.show()
plt.close()
```
